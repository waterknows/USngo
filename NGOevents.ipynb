```python
import pandas as pd
import re
from urllib.request import urlopen
from urllib.request import Request
from bs4 import BeautifulSoup
```

```python
link='https://blog.everyaction.com/2016-best-nonprofit-conferences-calendar'
```

```python
html = urlopen(link).read().decode('utf-8')
```

```python
soup=BeautifulSoup(html, "lxml")
```

```python
table=soup.find('tbody')
```

```python
table
```

```python
name=[]
url=[]
date=[]
city=[]
fee=[]
year=[]
description=[]
state=[]
category=[]

for each in table.find_all('tr'):
    nam=each.find('p', attrs={'class':'namehide'}).get_text()
    ur=each.find('a',attrs={'target':'blank'})['href']
    dat=each.find('div', attrs={'class':'date'}).get_text().replace('\n','')
    cit=each.find('div',attrs={'class':'info'}).get_text().split(' / ')[1].replace('\n','')
    fe=each.find('div',attrs={'class':'info'}).get_text().split(' / ')[2].replace('\n','')
    yea=each.find('div', attrs={'class':'year'}).get_text()
    desc=each.find('div', attrs={'class':'description'}).get_text().replace('\n','')
    stat=each.find_all('td', attrs={'class':'hide'})[0].get_text()
    cate=each.find_all('td', attrs={'class':'hide'})[1].get_text()
    name.append(nam)
    url.append(ur)
    date.append(dat)
    city.append(cit)
    fee.append(fe)
    year.append(yea)
    description.append(desc)
    state.append(stat)
    category.append(cate)
```

```python
df=pd.DataFrame({'URL':url,'Description':description,'Name':name,'Date':date,'Year':year,'City':city,'State':state,'Fee':fee,'Category':category})
```
