```python
import pandas as pd
import re
from urllib.request import urlopen
from urllib.request import Request
from bs4 import BeautifulSoup
```

```python
link='https://blog.everyaction.com/2016-best-nonprofit-conferences-calendar'
```

```python
html = urlopen(link).read().decode('utf-8')
```

```python
soup=BeautifulSoup(html, "lxml")
```

```python
table=soup.find('table',attrs={'class':'sortable'})
```

```python
table
```

```python
description=[]
for each in table.find_all('div', attrs={'class':'description'}):
    description.append(each.get_text())
```

```python
name=[]
for each in table.find_all('p', attrs={'class':'namehide'}):
    name.append(each.get_text())
```

```python
date=[]
for each in table.find_all('div', attrs={'class':'date'}):
    date.append(each.get_text())
```

```python
year=[]
for each in table.find_all('div', attrs={'class':'year'}):
    year.append(each.get_text())
```

```python
state=[]
for each in table.find_all('td', attrs={'class':'hide'}):
    year.append(each.get_text())
```

```python
df=pd.DataFrame['description':description,'name':name,'date':date,'year':year]
```

```python
dff = {'description':description,'name':name,'date':date,'year':year}
```
